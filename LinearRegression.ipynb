```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datafram=pd.DataFrame({'X':[i for i in range(1, 8)],'Y':[1.5, 3.8, 6.7, 9.0, 11.2, 13.6, 16.0]})
datafram
```

```python
datafram['X2']=datafram['X']**2
datafram['XY']=datafram['X']*datafram['Y']
datafram
```

```python
class LinearRegression:
    
    def __init__(self,datafram):
        self.variable=datafram
        self.sum_XY=datafram['XY'].sum()
        self.sum_X=datafram['X'].sum()
        self.sum_Y=datafram['Y'].sum()
        self.sum_X2=datafram['X2'].sum()
        self.n=len(datafram)
        
    def m_value(self):
        m_value_var=((self.n*self.sum_XY) - (self.sum_X * self.sum_Y))/((self.n*self.sum_X2) - (self.sum_X ** 2))
        return m_value_var

    def b_value(self):
        b_value_var=((self.sum_Y*self.sum_X2)-(self.sum_X*self.sum_XY))/((self.n*self.sum_X2)-(self.sum_X**2))
        return b_value_var
    
    def Y_predvalues(self,m_value_var,b_value_var,datafram):
        Y_pred_values=[((m_value_var*i)+b_value_var) for i in datafram['X']]
        datafram['Y_pred_values']=Y_pred_values
        return datafram
    
    def SSE_values(self,datafram):
        SSE=0
        for i,j in zip(datafram['Y_pred_values'],datafram['Y']):
            SSE+=(i-j)**2
        return SSE
    
    def graphplot(self):
        plt.scatter(datafram['X'], datafram['Y'])
        plt.plot(datafram['X'], datafram['Y_pred_values'], c='r')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.show()
       
# object of LinearRegression     
obj=LinearRegression(datafram)
m=obj.m_value()
b=obj.b_value()
print("The Slope of the graph is: ",m)
print("The intercept of the graph is: ",b)


df=obj.Y_predvalues(m,b,datafram)
datafram=df

print("Our Dataframe is:\n")
print(datafram)


sse=obj.SSE_values(datafram)
print("\n\nSum of Squared Errors: ",sse,"\n\n")

print("y=mx+c--- Linear Regression Graph")
obj.graphplot()
```
